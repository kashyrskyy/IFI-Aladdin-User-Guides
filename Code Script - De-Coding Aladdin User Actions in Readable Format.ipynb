```python
import json
from IPython.display import display, Markdown

# Load the JSON file
with open("exported_sample.json", "r") as file:
    data = json.load(file)

# Function to extract information and other modifications from nested json string
def get_json_info(json_str):
    try:
        parsed_data = json.loads(json_str)
        action_name = parsed_data.get('name', 'Unknown')
        modifications = {k: v for k, v in parsed_data.items() if k not in ['name', 'timestamp']}
        
        # If the action is 'Add', fetch the element type from 'addedElement' key
        if action_name == "Add" and 'addedElement' in parsed_data:
            added_element_type = parsed_data['addedElement'].pop('type', '')  # Remove 'type' key from modifications
            action_name += f" {added_element_type}"
            # Replace 'addedElement' key with its values
            modifications.update(parsed_data['addedElement'])
            modifications.pop('addedElement', None)  # Remove the original 'addedElement'
        
        # If the action is 'Delete', fetch the element type from 'deletedElements' key
        if action_name == "Delete" and 'deletedElements' in parsed_data and isinstance(parsed_data['deletedElements'], list) and len(parsed_data['deletedElements']) > 0:
            deleted_element_type = parsed_data['deletedElements'][0].pop('type', '')
            action_name += f" {deleted_element_type}"
            modifications.update(parsed_data['deletedElements'][0])
            modifications.pop('deletedElements', None)  # Remove the original 'deletedElements'
        
        # If the action is 'Paste by Key', fetch the element type from 'pasteElement' key
        if action_name == "Paste by Key" and 'pastedElements' in parsed_data:
            paste_element_type = parsed_data['pastedElements'][0].pop('type', '')
            action_name += f" {paste_element_type}"
            modifications.update(parsed_data['pastedElements'][0])
            modifications.pop('pastedElements', None)  # Remove the original 'pastedElements'

        # If the action is 'Cut', fetch the element type from 'deletedElements' key
        if action_name == "Cut" and 'deletedElements' in parsed_data and isinstance(parsed_data['deletedElements'], list) and len(parsed_data['deletedElements']) > 0:
            deleted_element_type = parsed_data['deletedElements'][0].pop('type', '')
            action_name += f" {deleted_element_type}"
            modifications.update(parsed_data['deletedElements'][0])
            modifications.pop('deletedElements', None)  # Remove the original 'deletedElements'

        # If the action is 'Paste to Point', fetch the element type from 'pastedElements' key
        if action_name == "Paste to Point" and 'pastedElements' in parsed_data:
            paste_element_type = parsed_data['pastedElements'][0].pop('type', '')
            action_name += f" {paste_element_type}"
            modifications.update(parsed_data['pastedElements'][0])
            modifications.pop('pastedElements', None)  # Remove the original 'pastedElements'
            
        # Handle "Move" action
        if action_name == "Move" and 'movedElementType' in parsed_data:
            moved_element_type = parsed_data['movedElementType']
            action_name += f" {moved_element_type}"
            # Remove 'movedElementType' key from modifications
            modifications.pop('movedElementType', None)
        
        # Handle "Copy" action
        if action_name == "Copy" and 'elementType' in parsed_data:
            copied_element_type = parsed_data['elementType']
            action_name += f" {copied_element_type}"
            # Remove 'elementType' key from modifications
            modifications.pop('elementType', None)

        return action_name, modifications
    except:
        return 'Unknown', {}    
    
# Process the data
output = []

user_counter = 1  # This counter will be used to rename users to User 1, User 2, etc.
for user_id, timestamps in data.items():
    user_output = f"### User {user_counter} Actions Log:"  # Use the counter instead of actual user_id
    action_list = []
    for timestamp, action_data in timestamps.items():
        action_name, action_modifications = get_json_info(action_data.get('action', {}))
        
        modifications_list = [f"{k}: {v}" for k, v in action_modifications.items() if k != "addedElement"]
        mods = ""
        if modifications_list:
            mods = "\n    - " + "\n    - ".join(modifications_list)
        
        action_description = f"- **At {timestamp}:** *{action_name}*{mods}"
        
        action_list.append(action_description)
    
    user_output += "\n" + "\n".join(action_list)
    output.append(user_output)

    user_counter += 1  # Increment the counter for the next user

# Display the result using Markdown
for o in output:
    display(Markdown(o))
    display(Markdown("---"))
```
